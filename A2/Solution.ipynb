#### Q6. Remove books and users with less than 20 rating scores from the utility matrix by using for modifying the following codes.

#### Q7. Partition the data set for cross validating the performance on predicting rating on Book 8010. Use 80% of the users as training and the rest as testing. What are the dimensions of the training set and the testing set? What are the mean rating of Book 8010 in the training set and the testing set? (Hint: use dense_matrix[8010].mean() method to calculate the means)

```python
from sklearn.cross_validation import train_test_split

# create a data frame for the predictors
df_x = dense_matrix[[col for col in dense_matrix.columns if col != 8010]]
# print df_x.shape

# create outcome
df_y = dense_matrix[[8010]]
# print df_y.shape


train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.2, random_state=0)
df_train_x = pd.DataFrame(train_x, columns=df_x.columns)
df_test_x = pd.DataFrame(test_x, columns=df_x.columns)
df_train_y = pd.DataFrame(train_y, columns=[8010])
df_test_y = pd.DataFrame(test_y, columns=[8010])
print "Training dimensions: "
print df_train_x.shape
print df_train_y.shape
print "Testing dimensions: "
print df_test_x.shape
print df_test_y.shape

# print 
# print "class counts"
# print df_train_y[151].value_counts()
# print df_test_y[151].value_counts()
```

```python
print "Mean rating of 8010 in training data: %f" % df_train_y.mean()
print "Mean rating of 8010 in testing data: %f" % df_test_y.mean()
```

#### Q8. Define a function that implements user-based collaborative filtering to predict the ratings on Book 8010 based on the top 10 similar users. What is the predicted rating for User 23? What is the user’s actual rating?

```python
# specify the number of similar users to retrieve
k = 10

def user_based_predict(user_number):
    # retrieve the top k similar users

    # copy from all the training predictors
    df_sim = df_train_x.copy()

    # for each user, calculate the distance between this user and the active user
    df_sim["distance"] = df_sim.apply(lambda x: euclidean(df_test_x.loc[user_number], x), axis=1)
    # create a new data frame to store the top k similar users
    df_sim_users = df_sim.loc[df_sim.sort("distance").head(k).index]

    # calculate these similar users' rating on 8010, weighted by distance
    df_sim_users["weighed_d"] = map(lambda x: df_sim_users.loc[x]["distance"]*df_train_y.loc[x][8010], df_sim_users.index)
    predicted = df_sim_users["weighed_d"].sum()/df_sim_users["distance"].sum()
    return predicted
    
print "Predicted rating for 23: " , user_based_predict(23)
print "Actual rating for 23: " , dense_matrix.loc[23][8010]
```

#### Q9. What is the mean squared error for this function on predicting the rating of all users in the testing set on Book 8010? (This may take more than 10 minutes, depending on your computer.)

```python
# predict all testing data
%time
pred_testing = map(user_based_predict, df_test_x.index)
print "Number of ratings predicted: %d" % len(pred_testing)
```

```python
from sklearn.metrics import mean_squared_error
print "Mean Squared Error: " , mean_squared_error(pred_testing, df_test_y[8010])
```
